<h1>Getting Started with TFDS</h1>

Example: Fashion MNIST

```python
import tensorflow as tf
import tensorflow_datasets as tfds
mnist_data = tfds.load('fashion_mnist')
for item in mnist_data:
    print(item)
```

Load splits into a dataset containing actual data.

```python
mnist_train = tfds.load(name='fashion_mnist', split='train')
assert isinstance(mnist_train, tf.data.Dataset)
print(type(mnist_train))
```

Call take(1) to get the first element (record) of the dataset.

```python
for item in mnist_train.take(1):
    print(type(item))
    print(item.keys())
```

Inspect a value in a dataset.

```python
for item in mnist_train.take(1):
    print(type(item))
    print(item.keys())
    print(item['image'])
    print(item['label'])
```

tf.Tensor(2, shape=(), dtype=int64) indicates that this image is class 2 in the MNIST dataset.

Using with_info parameter when loading the dataset, you can get the info about the dataset.

```python
mnist_train, info = tfds.load(name='fashion_mnist', with_info=True)
print(info)
```

<h1>Using TFDS with Keras Models</h1>

Same format of data was available in the Keras datasts, with one modification - the shape in TFDS is (28, 28, 1), whereas in the Keras datasets it was (28, 28).

Example: Horses or Humans dataset. 
Code to train a model with it.

```python
import tensorflow as tf
import tensorflow_datasets as tfds

data = tfds.load('horses_or_humans', split='train', as_supervised=True)

train_batches = data.shuffle(100).batch(10)

model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(16, (3, 3), activation='relu', input_shape=(300, 300, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

history = model.fit(train_batches, epochs=10)
```

Vadlidation of the model while training by loading a separate validation set from TFDS.

```python
val_data = tfds.load('horses_or_humans', split='test', as_supervised=True)
```

```python
# Batching the validation data.

validation_batches = val_data.batch(32)
```

```python
history = model.fit(train_batches, epochs=10, validation_data=validation_batches, validation_steps=1)
```

<h1>ETL Process for Managing Data in TensorFlow.</h1>

Consider Horses or Humans classifier.

```python
import tensorflow as tf
import tensorflow_datasets as tfds
import tensorflow_addons as tfa

# Model definition start #
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(16, (3, 3), activation='relu', input_shape=(300, 300, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
# Model definition end #

# Extract Phase Start #
data = tfds.load('horses_or_humans', split='train', as_supervised=True)
val_data = tfds.load('horses_or_humans', split='test', as_supervised=True)
# Extract Phase End #

# Transform phase start #
def augmentimages(image, label):
    image = tf.cast(image, tf.float32)
    image = (image / 255)
    image = tf.image.random_flip_left_right(image)
    image = tfa.image.rotate(image, 40, interpolation='NEAREST')
    return image, label

train = data.map(augmentimages)
train_batches = data.shuffle(100).batch(32)
validation_batches = val_data.batch(32)
# Transform phase end #

# Load Phase Start #
history = model.fit(train_batches, epochs=10, validation_data=validation_batches, validation_steps=1)
# Load Phase End #
```

<h1>Parallelizing ETL to improve training performance.</h1>

Example dataset: Dogs vs. Cats dataset.

```python
# Accessing raw files that were downloaded.
# Create a list of these files ans use tf.Data.Dataset.list_files to load them.
file_pattern = f'path/to/data/*.tfrecord'
files = tf.data.Dataset.list_files(file_pattern)
```

```python
# Loaded into dataset using files.interleave

train_dataset = files.interleave(tf.data.TFRecordDataset, cycle_length=4. num_parallel_calls=tf.data.experimental.AUTOTUNE)
```

"cycle_length" parameter specifies the number of input elements that are processed concurrently.
'cycle_length = 4', this process will be handling four records at a time.

"num_parallel_calls" parameter specifies the number of parallel calls to execute.

tf.data.experimental.AUTOTUNE, the value is set dynamically based on available CPUs.

Example: If num_parallel_calls is set to 6 after autotuning and cycle_length is 4, you will have 6 separate threads, each running 4 records at a time.
