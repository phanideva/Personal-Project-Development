```python
import pygame
import random

# Initialize Pygame
pygame.init()

# Set up display
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 600

screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption('Danny: The Snake')

# Set colors
WHITE = (255, 255, 255)
YELLOW = (255, 255, 102)
BLACK = (0, 0, 0)
RED = (213, 50, 80)
GREEN = (0, 255, 0)
BLUE = (50, 153, 213)
ENEMY_COLOR = (150, 0, 150)

# Set game clock
clock = pygame.time.Clock()
SNAKE_BLOCK = 20
SNAKE_SPEED = 15

# Set fonts
font_style = pygame.font.SysFont("bahnschrift", 25)
score_font = pygame.font.SysFont("comicsansms", 35)

# Load and play background music
pygame.mixer.music.load("background.mp3")
pygame.mixer.music.play(-1)  # Loop the music indefinitely

def our_snake(snake_block, snake_list, color):
    for x in snake_list:
        pygame.draw.rect(screen, color, [x[0], x[1], snake_block, snake_block])

def message(msg, color, y_displace=0):
    mesg = font_style.render(msg, True, color)
    screen.blit(mesg, [SCREEN_WIDTH / 6, SCREEN_HEIGHT / 3 + y_displace])

def draw_score(score):
    value = score_font.render("Score: " + str(score), True, YELLOW)
    screen.blit(value, [0, 0])

def draw_lives(lives):
    value = score_font.render("Lives: " + str(lives), True, YELLOW)
    screen.blit(value, [0, 40])

def generate_food(snake_list, enemy_snakes):
    while True:
        foodx = round(random.randrange(0, SCREEN_WIDTH - SNAKE_BLOCK) / SNAKE_BLOCK) * SNAKE_BLOCK
        foody = round(random.randrange(0, SCREEN_HEIGHT - SNAKE_BLOCK) / SNAKE_BLOCK) * SNAKE_BLOCK
        if [foodx, foody] not in snake_list and all([foodx, foody] not in enemy for enemy in enemy_snakes):
            return foodx, foody

def gameLoop():
    game_over = False
    game_close = False

    x1 = SCREEN_WIDTH / 2
    y1 = SCREEN_HEIGHT / 2

    x1_change = 0
    y1_change = 0

    snake_List = []
    Length_of_snake = 1

    enemy_snakes = [[[100, 100]], [[700, 500]], [[300, 300]]]
    enemy_directions = [[SNAKE_BLOCK, 0], [0, -SNAKE_BLOCK], [-SNAKE_BLOCK, 0]]

    foodx, foody = generate_food(snake_List, enemy_snakes)

    score = 0
    lives = 5

    while not game_over:

        while game_close:
            screen.fill(BLUE)
            message("You Lost a Life! Press C to Continue or Q to Quit", RED, -50)
            message(f"Score: {score}", YELLOW, 50)
            pygame.display.update()

            for event in pygame.event.get():
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_q:
                        game_over = True
                        game_close = False
                    if event.key == pygame.K_c:
                        game_close = False
                        x1 = SCREEN_WIDTH / 2
                        y1 = SCREEN_HEIGHT / 2
                        x1_change = 0
                        y1_change = 0
                        snake_List = []
                        Length_of_snake = 1
                        foodx, foody = generate_food(snake_List, enemy_snakes)

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    x1_change = -SNAKE_BLOCK
                    y1_change = 0
                elif event.key == pygame.K_RIGHT:
                    x1_change = SNAKE_BLOCK
                    y1_change = 0
                elif event.key == pygame.K_UP:
                    y1_change = -SNAKE_BLOCK
                    x1_change = 0
                elif event.key == pygame.K_DOWN:
                    y1_change = SNAKE_BLOCK
                    x1_change = 0

        x1 += x1_change
        y1 += y1_change
        screen.fill(WHITE)
        pygame.draw.rect(screen, GREEN, [foodx, foody, SNAKE_BLOCK, SNAKE_BLOCK])
        snake_Head = []
        snake_Head.append(x1)
        snake_Head.append(y1)
        snake_List.append(snake_Head)
        if len(snake_List) > Length_of_snake:
            del snake_List[0]

        for x in snake_List[:-1]:
            if x == snake_Head:
                lives -= 1
                if lives == 0:
                    game_close = True
                else:
                    game_close = True

        for i, enemy in enumerate(enemy_snakes):
            if len(enemy) > 1:
                del enemy[0]

            head = enemy[-1][:]
            direction = enemy_directions[i]
            head[0] += direction[0]
            head[1] += direction[1]

            if head[0] >= SCREEN_WIDTH or head[0] < 0 or head[1] >= SCREEN_HEIGHT or head[1] < 0:
                direction[0] = -direction[0]
                direction[1] = -direction[1]

            enemy.append(head)

            if snake_Head == head:
                lives -= 1
                if lives == 0:
                    game_close = True
                else:
                    game_close = True

            for segment in enemy[:-1]:
                if snake_Head == segment:
                    lives -= 1
                    if lives == 0:
                        game_close = True
                    else:
                        game_close = True

        if x1 == foodx and y1 == foody:
            foodx, foody = generate_food(snake_List, enemy_snakes)
            Length_of_snake += 1
            score += 1

        our_snake(SNAKE_BLOCK, snake_List, BLACK)
        for enemy in enemy_snakes:
            our_snake(SNAKE_BLOCK, enemy, ENEMY_COLOR)
        
        draw_score(score)
        draw_lives(lives)
        pygame.display.update()

        if x1 >= SCREEN_WIDTH or x1 < 0 or y1 >= SCREEN_HEIGHT or y1 < 0:
            lives -= 1
            if lives == 0:
                game_close = True
            else:
                game_close = True

        clock.tick(SNAKE_SPEED)

    pygame.quit()
    quit()

def start_screen():
    screen.fill(BLUE)
    message("Danny: The Snake", GREEN, -100)
    message("Press P to Play", WHITE, 0)
    pygame.display.update()
    
    waiting = True
    while waiting:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_p:
                    waiting = False
                    gameLoop()

while True:
    start_screen()
```

```python
import pygame
import random

# Initialize Pygame
pygame.init()

# Set up display
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 600

screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption('Danny: The Snake')

# Set colors
WHITE = (255, 255, 255)
YELLOW = (255, 255, 102)
BLACK = (0, 0, 0)
RED = (213, 50, 80)
GREEN = (0, 255, 0)
BLUE = (50, 153, 213)
ENEMY_COLOR = (150, 0, 150)
APPLE_COLOR = (255, 0, 0)
SNAKE_COLOR = (0, 128, 0)

# Set game clock
clock = pygame.time.Clock()
SNAKE_BLOCK = 20
SNAKE_SPEED = 15

# Set fonts
font_style = pygame.font.SysFont("bahnschrift", 35)
score_font = pygame.font.SysFont("comicsansms", 45)
level_font = pygame.font.SysFont("comicsansms", 35)

# Load and play background music
pygame.mixer.music.load("background.mp3")
pygame.mixer.music.play(-1)  # Loop the music indefinitely

def draw_snake(snake_list, color):
    for x in snake_list:
        pygame.draw.rect(screen, color, [x[0], x[1], SNAKE_BLOCK, SNAKE_BLOCK])
        pygame.draw.rect(screen, BLACK, [x[0], x[1], SNAKE_BLOCK, SNAKE_BLOCK], 1)  # Add border for better visibility

def message(msg, color, font, y_displace=0):
    mesg = font.render(msg, True, color)
    screen.blit(mesg, [SCREEN_WIDTH / 6, SCREEN_HEIGHT / 3 + y_displace])

def draw_score(score):
    value = score_font.render("Score: " + str(score), True, YELLOW)
    screen.blit(value, [10, 10])

def draw_lives(lives):
    value = score_font.render("Lives: " + str(lives), True, YELLOW)
    screen.blit(value, [10, 50])

def draw_level(level):
    value = level_font.render("Level: " + str(level), True, YELLOW)
    screen.blit(value, [10, 90])

def generate_food(snake_list, enemy_snakes):
    while True:
        foodx = round(random.randrange(0, SCREEN_WIDTH - SNAKE_BLOCK) / SNAKE_BLOCK) * SNAKE_BLOCK
        foody = round(random.randrange(0, SCREEN_HEIGHT - SNAKE_BLOCK) / SNAKE_BLOCK) * SNAKE_BLOCK
        if [foodx, foody] not in snake_list and all([foodx, foody] not in enemy for enemy in enemy_snakes):
            return foodx, foody

def gameLoop():
    game_over = False
    game_close = False

    x1 = SCREEN_WIDTH / 2
    y1 = SCREEN_HEIGHT / 2

    x1_change = 0
    y1_change = 0
    direction = 'RIGHT'

    snake_List = [[x1, y1]]
    Length_of_snake = 1

    enemy_snakes = [[[100, 100]], [[700, 500]], [[300, 300]]]
    enemy_directions = [[SNAKE_BLOCK, 0], [0, -SNAKE_BLOCK], [-SNAKE_BLOCK, 0]]

    foodx, foody = generate_food(snake_List, enemy_snakes)

    score = 0
    lives = 5
    level = 1

    while not game_over:

        while game_close:
            screen.fill(BLUE)
            message("You Lost a Life! Press C to Continue or Q to Quit", RED, font_style, -50)
            message(f"Score: {score}", YELLOW, font_style, 50)
            pygame.display.update()

            for event in pygame.event.get():
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_q:
                        game_over = True
                        game_close = False
                    if event.key == pygame.K_c:
                        game_close = False
                        x1 = SCREEN_WIDTH / 2
                        y1 = SCREEN_HEIGHT / 2
                        x1_change = 0
                        y1_change = 0
                        snake_List = [[x1, y1]]
                        Length_of_snake = 1
                        foodx, foody = generate_food(snake_List, enemy_snakes)
                        break

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT and direction != 'RIGHT':
                    x1_change = -SNAKE_BLOCK
                    y1_change = 0
                    direction = 'LEFT'
                elif event.key == pygame.K_RIGHT and direction != 'LEFT':
                    x1_change = SNAKE_BLOCK
                    y1_change = 0
                    direction = 'RIGHT'
                elif event.key == pygame.K_UP and direction != 'DOWN':
                    y1_change = -SNAKE_BLOCK
                    x1_change = 0
                    direction = 'UP'
                elif event.key == pygame.K_DOWN and direction != 'UP':
                    y1_change = SNAKE_BLOCK
                    x1_change = 0
                    direction = 'DOWN'

        if x1 >= SCREEN_WIDTH or x1 < 0 or y1 >= SCREEN_HEIGHT or y1 < 0:
            lives -= 1
            if lives == 0:
                game_close = True
            else:
                game_close = False
                x1 = SCREEN_WIDTH / 2
                y1 = SCREEN_HEIGHT / 2
                x1_change = 0
                y1_change = 0
                snake_List = [[x1, y1]]
                Length_of_snake = 1
                foodx, foody = generate_food(snake_List, enemy_snakes)

        x1 += x1_change
        y1 += y1_change
        screen.fill(WHITE)
        pygame.draw.rect(screen, APPLE_COLOR, [foodx, foody, SNAKE_BLOCK, SNAKE_BLOCK])

        snake_Head = [x1, y1]
        snake_List.append(snake_Head)
        if len(snake_List) > Length_of_snake:
            del snake_List[0]

        for segment in snake_List[:-1]:
            if segment == snake_Head:
                lives -= 1
                if lives == 0:
                    game_close = True
                else:
                    game_close = False

        for i, enemy in enumerate(enemy_snakes):
            if len(enemy) > 1:
                del enemy[0]

            head = enemy[-1][:]
            direction = enemy_directions[i]
            head[0] += direction[0]
            head[1] += direction[1]

            if head[0] >= SCREEN_WIDTH or head[0] < 0 or head[1] >= SCREEN_HEIGHT or head[1] < 0:
                direction[0] = -direction[0]
                direction[1] = -direction[1]

            enemy.append(head)

            if snake_Head == head:
                lives -= 1
                if lives == 0:
                    game_close = True
                else:
                    game_close = False

            for segment in enemy[:-1]:
                if snake_Head == segment:
                    lives -= 1
                    if lives == 0:
                        game_close = True
                    else:
                        game_close = False

        if x1 == foodx and y1 == foody:
            foodx, foody = generate_food(snake_List, enemy_snakes)
            Length_of_snake += 1
            score += 1
            if score % 5 == 0:
                level += 1
                global SNAKE_SPEED
                SNAKE_SPEED += 5

        draw_snake(snake_List, SNAKE_COLOR)
        for enemy in enemy_snakes:
            draw_snake(enemy, ENEMY_COLOR)

        draw_score(score)
        draw_lives(lives)
        draw_level(level)
        pygame.display.update()

        clock.tick(SNAKE_SPEED)

    pygame.quit()
    quit()

def start_screen():
    screen.fill(BLUE)
    message("Danny: The Snake", GREEN, font_style, -100)
    message("Press P to Play", WHITE, font_style, 0)
    pygame.display.update()
    
    waiting = True
    while waiting:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_p:
                    waiting = False
                    gameLoop()

while True:
    start_screen()
```
