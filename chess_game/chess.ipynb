```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Screen dimensions
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 800
SQUARE_SIZE = SCREEN_WIDTH // 8

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
GREEN = (0, 128, 0)
BLUE = (0, 0, 255)

# Initialize screen
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Chess Game")

# Load piece images
pieces = {
    "bR": pygame.image.load("images/bR.png"),
    "bN": pygame.image.load("images/bN.png"),
    "bB": pygame.image.load("images/bB.png"),
    "bQ": pygame.image.load("images/bQ.png"),
    "bK": pygame.image.load("images/bK.png"),
    "bP": pygame.image.load("images/bP.png"),
    "wR": pygame.image.load("images/wR.png"),
    "wN": pygame.image.load("images/wN.png"),
    "wB": pygame.image.load("images/wB.png"),
    "wQ": pygame.image.load("images/wQ.png"),
    "wK": pygame.image.load("images/wK.png"),
    "wP": pygame.image.load("images/wP.png"),
}

# Chess board setup
board = [
    ["bR", "bN", "bB", "bQ", "bK", "bB", "bN", "bR"],
    ["bP", "bP", "bP", "bP", "bP", "bP", "bP", "bP"],
    ["", "", "", "", "", "", "", ""],
    ["", "", "", "", "", "", "", ""],
    ["", "", "", "", "", "", "", ""],
    ["", "", "", "", "", "", "", ""],
    ["wP", "wP", "wP", "wP", "wP", "wP", "wP", "wP"],
    ["wR", "wN", "wB", "wQ", "wK", "wB", "wN", "wR"],
]

def draw_board():
    for row in range(8):
        for col in range(8):
            color = WHITE if (row + col) % 2 == 0 else GREEN
            pygame.draw.rect(screen, color, (col * SQUARE_SIZE, row * SQUARE_SIZE, SQUARE_SIZE, SQUARE_SIZE))

def draw_pieces():
    for row in range(8):
        for col in range(8):
            piece = board[row][col]
            if piece != "":
                screen.blit(pieces[piece], (col * SQUARE_SIZE, row * SQUARE_SIZE))

def get_square_under_mouse():
    mouse_pos = pygame.mouse.get_pos()
    x, y = mouse_pos[0] // SQUARE_SIZE, mouse_pos[1] // SQUARE_SIZE
    return (x, y)

def move_piece(start_pos, end_pos):
    piece = board[start_pos[1]][start_pos[0]]
    board[start_pos[1]][start_pos[0]] = ""
    board[end_pos[1]][end_pos[0]] = piece

def draw_text(text, font, color, surface, x, y):
    text_obj = font.render(text, True, color)
    text_rect = text_obj.get_rect()
    text_rect.center = (x, y)
    surface.blit(text_obj, text_rect)

def start_screen():
    screen.fill(BLACK)
    font = pygame.font.Font(None, 74)
    draw_text("Chess Game", font, WHITE, screen, SCREEN_WIDTH // 2, SCREEN_HEIGHT // 2 - 100)
    draw_text("Press any key to start", font, WHITE, screen, SCREEN_WIDTH // 2, SCREEN_HEIGHT // 2 + 100)
    pygame.display.flip()
    waiting = True
    while waiting:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
            if event.type == pygame.KEYDOWN:
                waiting = False

def game_over_screen(winner):
    screen.fill(BLACK)
    font = pygame.font.Font(None, 74)
    draw_text(f"Game Over: {winner} Wins", font, WHITE, screen, SCREEN_WIDTH // 2, SCREEN_HEIGHT // 2 - 100)
    draw_text("Press any key to restart", font, WHITE, screen, SCREEN_WIDTH // 2, SCREEN_HEIGHT // 2 + 100)
    pygame.display.flip()
    waiting = True
    while waiting:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
            if event.type == pygame.KEYDOWN:
                waiting = False

def is_in_check(board, color):
    king_pos = None
    for row in range(8):
        for col in range(8):
            if board[row][col] == f"{color}K":
                king_pos = (col, row)
                break
    if not king_pos:
        return False

    opponent_color = "w" if color == "b" else "b"
    for row in range(8):
        for col in range(8):
            piece = board[row][col]
            if piece.startswith(opponent_color):
                if can_move(board, piece, (col, row), king_pos):
                    return True
    return False

def can_move(board, piece, start, end):
    piece_type = piece[1]
    if piece_type == "P":
        return can_move_pawn(board, piece, start, end)
    # Implement other piece movement logic here (rook, knight, bishop, queen, king)
    return False

def can_move_pawn(board, piece, start, end):
    direction = -1 if piece.startswith("w") else 1
    if start[0] == end[0] and start[1] + direction == end[1] and board[end[1]][end[0]] == "":
        return True
    return False

def is_checkmate(board, color):
    for row in range(8):
        for col in range(8):
            if board[row][col].startswith(color):
                for r in range(8):
                    for c in range(8):
                        if can_move(board, board[row][col], (col, row), (c, r)):
                            new_board = [row[:] for row in board]
                            move_piece((col, row), (c, r))
                            if not is_in_check(new_board, color):
                                return False
    return True

selected_square = None
turn = "w"

# Main loop
running = True
while running:
    start_screen()
    
    game_over = False
    while not game_over:
        draw_board()
        draw_pieces()
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True
                running = False
            elif event.type == pygame.MOUSEBUTTONDOWN:
                square = get_square_under_mouse()
                if selected_square:
                    move_piece(selected_square, square)
                    if is_in_check(board, "w" if turn == "b" else "b"):
                        if is_checkmate(board, "w" if turn == "b" else "b"):
                            game_over = True
                            game_over_screen("White" if turn == "w" else "Black")
                    turn = "b" if turn == "w" else "w"
                    selected_square = None
                else:
                    selected_square = square
        
        pygame.display.flip()
    
    game_over_screen("White" if turn == "b" else "Black")

pygame.quit()
sys.exit()
```
